```python
%sql
DROP TABLE IF EXISTS monschema1.orders_bronze;
CREATE OR REPLACE TABLE monschema1.orders_bronze AS 
SELECT *, current_timestamp() AS processing_time, _metadata.file_name AS source_file
FROM read_files(
  "/Volumes/workspace/monschema1/my_volume/orders", 
  format => 'json',
  schema => 'customer_id INT, order_id INT, notification STRING, order_timestamp TIMESTAMP'
);


CREATE OR REPLACE TABLE monschema1.orders_silver
AS SELECT order_id, 
timestamp(order_timestamp) AS order_timestamp,
customer_id, 
notification FROM monschema1.orders_bronze;


CREATE OR REPLACE VIEW monschema1.orders_by_date_vw
AS select date(order_timestamp) as order_date,count(*) as total_daily_orders from monschema1.orders_silver group by date(order_timestamp); 
 
```
